# @데코는 사용자 정의로 만들어진 메서드를 호출하는 방식이고 그 메서드 매개변수에 함수를 넣는 방식인가?
# def deco() 메서드에 매개변수가 꼭 있어야 함

```python
# 데코레이터: 기존 함수를 고치지 않고, 기능을 덧붙히는 방법


def deco(fun):
    def wrap():
        print('실행 전')
        fun()
        print('실행 후')
    return wrap



#def hello():
#    print("안녕하세요")
#new_fun=deco(hello)
#new_fun()



@deco
def hello():
     print("안녕하세요")

hello()
```
