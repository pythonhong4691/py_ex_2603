```python
# 데이터 형식 모듈
# JSON 형식으로 데이터를 변환
# JSON: ("Key: Values") 형식으로 저장
# API: 프로그램끼리 대화하는 방법 (규칙)

import json

data={
    "name":"houng",
    "age":25
}
print(json.dumps(data))
```

```python
 #정규표현식 모듈 (파싱)
# 문자열 패턴 검사 모듈
# 형식 확인(아이디, 이메일 등)에 사용

import re

# 문자열이 hi로 시작하는지 확인
print(bool(re.match('hi','hi python')))


```

```python
# CSV 파일 모듈
# CSV(엑셀) 파일을 읽고 쓰는 모듈
# 데이터 저장/분석 기초에 사용

import csv

# CSV 파일에 한 줄 데이터를 저장
with open('test.csv','w',newline='') as f:
    csv.writer(f).writerow(['name','score'])




```
