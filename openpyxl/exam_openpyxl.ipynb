```python
import openpyxl  # 모든 기능을 불러옴
from openpyxl import Workbook
from openpyxl import load_workbook
from openpyxl import load_workbook
# 기존 엑셀 파일 열기
#w2 = load_workbook('test.xlsx')

# 버전 확인
print(openpyxl.__version__)


# === 엑셀 파일 생성, 기본 시트 선택, 셀 값 넣기 ===

# 새 엑셀 파일을 만들겠다
#w1=Workbook()

# 만들어진 엑셀 파일에서 기본 시트를 선택
#ws=w1.active

# 지금 만든 엑셀 파일을 실제 파일로 저장
#w1.save('test.xlsx')

# 기존에 저장된 엑셀 파일을 다시 열어서 수정할 수 있게 함
w2=load_workbook('test.xlsx')

# 열어 온 엑셀 파일에서 첫 번째 시트 선택
ws2=w2.active

#셀에 값 입력하기
# A열 1행에 Hello 값 넣기
#ws2['C3']='HIIIIII'
#w2.save('test.xlsx')

a='python'
# ws2['D5':'F8'] : D5부터 F8셀 까지의 범위를 선택 
for row in ws2['D5':'F8']: # 행 단위로 반복
    for cell in row: # 각 행 안의 셀들을 하나씩 꺼내서 python값 입력
        cell.value=a


w2.save('test.xlsx')

```

```python
from openpyxl import *

# == 딕셔너리로 값 생성 후 액셀에 넣기 ==

item1={
    '품목': '바나나',
    '가격': '3000',
    '수량': 1,
    '날짜':'2026년 3월 10일'
}
item2={
    '품목': '책',
    '가격': '15000',
    '수량': 20,
    '날짜':'2026년 3월 11일'
}
item3={
    '품목': '커피',
    '가격': '3500',
    '수량': 10,
    '날짜':'2026년 3월 12일'
}
item4={
    '품목': '신발',
    '가격': '50000',
    '수량': 4,
    '날짜':'2026년 3월 13일'
}

wb=Workbook()
ws=wb.active

ws['A1']='품목'
ws['B1']='가격'
ws['C1']='수량'
ws['D1']='날짜'

itmes=[item1,item2,item3,item4]

i=2

for item in itmes:
    ws['A'+str(i)]=item['품목']
    ws['B'+str(i)]=item['가격']
    ws['C'+str(i)]=item['수량']
    ws['D'+str(i)]=item['날짜']
    i+=1

wb.save('test111.xlsx')






```

```python
from openpyxl import *
from openpyxl.styles import PatternFill


# === 셀 배경 색 바꾸기 ===
wb=Workbook()
ws=wb.active

# 헤더
ws['A1']='이름'
ws['B1']='국어'
ws['C1']='영어'
ws['D1']='수학'

colors=PatternFill(fill_type='solid',fgColor='ed931c')

for cell in ws[1]: # 1행 전체
    cell.fill=colors

wb.save('bg_color.xlsx')



```

```python
from openpyxl import *
from openpyxl.chart import BarChart,Reference

# === 차트 만들기 ===

wb=Workbook()
ws=wb.active

# 예제 데이터 입력
ws.append(['Name','Score']) # 헤더 행
ws.append(['홍길동',90])
ws.append(['임꺽정',80])
ws.append(['홍길순',70])
ws.append(['임꺽순',85])

# BarChart 객체 생성 (막대 그래프)
chart=BarChart()

# 차트에 사용할 데이터 범위 지정
# min_col=2 -> B열(Score)
# min_row=1 -> 1행부터
# max_row=5 -> 5행까지
data= Reference(ws,min_col=2,min_row=1,max_row=5)

# 지정한 데이터를 차트에 추가
# 첫 행 Series name(계열 이름)으로 사용
chart.add_data(data,titles_from_data=True)


chart.title="Student Scores"

# 차트를 워크시트의 E5 위치에 추가
ws.add_chart(chart,"E5")


wb.save('chart_exam.xlsx')
```
