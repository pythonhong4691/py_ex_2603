```python
# 파이썬 유니코드 / 인코딩 핵심 정리


# 1. 문자열은 컴퓨터 내부에서 숫자(유니코드)로 저장
# -내부적으로 각 글자를 숫자로 바꿔 저장
# -그 숫자 체계를 유니코드(unicode)라고 한다.

text='파이썬'

print(text)      # 문자열 출력
print(len(text)) # 문자열 길이 출력

# 2. 문자 숫자 변환
# String : 문자열
# char : 문자

print(ord('가'))    # 문자를 유니코드 숫자로 변환
print(chr(44032))   # 숫자를 다시 문자로 변환

# 3. 문자열을 바이트로 변환 (encode)
# -encode : 문자열을 바이트(byte)로 바꾸는 작업
# 파일 저장이나 네트워크 전송 시 사용됨
print('-------------------------------')
encoded=text.encode('utf-8')
print(encoded)
print(type(encoded))

# 4. 바이트를 다시 문자열로 변환 (decode)
print('-------------------------------')
decoded=encoded.decode('utf-8')
print(decoded)
print(type(decoded))

# 5. 인코딩이 맞지 않으면 오류 발생

try:
    wrong=encoded.decode('cp949')   # utf-8로 하기로 했는데 cp949로 하면 오류
    print(wrong)
except UnicodeDecodeError:
    print("인코딩 방식이 맞지 않아 오류 발생")


# unicode: 문자 -> 번호
# UTF-8:   번호 -> 바이트 (한국어 인코드 방식)
# encode:  문자열 -> 바이트
# decode:  바이트 -> 문자열
```
