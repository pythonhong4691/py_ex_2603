```python
from selenium import webdriver
from selenium.webdriver.common.by import By

# 브라우저 실행 + 브라우저 접속
driver=webdriver.Chrome()
driver.get('https://quotes.toscrape.com/')


# 다음 페이지 이동 + 반복 수집


MAX_PAGES= 3    # 최대 몇 페이지까지 수집할 지 제한하는 상수 -> 무한 크롤링을 방지하기 위한 안전 장치
current_page=1  # 현재 페이지 번호를 추적하기 위한 변수



# Next 버튼 찾기
# "li.next a" : class='next'를 가진 li안의 a태그를 의미
# CSS_SELECTOR: CSS문법을 이용해서 HTML 요소를 찾는 방법
# Selenium에서 가장 많이 쓰이는 방식 중 하나

while(current_page <=MAX_PAGES):

    
    # 현재 수집 중인 페이지 번호 출력
    print(f"===={current_page}페이지 수집 중 ====")
    #새 페이지에서 다시 quote 블록 수집
    quote_blokcs=driver.find_elements(By.CLASS_NAME,'quote')


    for block in quote_blokcs:
        quote_text=block.find_element(By.CLASS_NAME,'text').text
        author_text=block.find_element(By.CLASS_NAME,'author').text
        print(f"{author_text}: {quote_text}\n")


    next_buttons=driver.find_elements(By.CSS_SELECTOR,'li.next a')
    if not next_buttons:
        print('다음 페이지가 없어 종료합니다.')
        break
    # 다음 페이지로 이동
    next_buttons[0].click()
    current_page +=1

    
# 브라우저 종료
driver.quit()



    




```
