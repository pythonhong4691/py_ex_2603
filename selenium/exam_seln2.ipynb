```python
from selenium import webdriver
from selenium.webdriver.common.by import By

# =============================================
# 1) 브라우저 실행 + 페이지 접속
# =============================================


driver=webdriver.Chrome()
#driver.get('https://www.naver.com/')
driver.get('https://quotes.toscrape.com/')

# =============================================
# 1) 태그 이름으로 요소 찾기(TAG_NAME)
#  HTML 태그 (div, span, a)으로 요소 찾기 -> 범위가 너무 많아서 주로 '연습용'으로 사용
# =============================================


div_elements=driver.find_elements(By.TAG_NAME,'div')
print('TAG_NAME으로 찾은 div 태그 개수: ',len(div_elements))

# =============================================
# 2) class로 요소 찾기(CLASS_NAME)
# HTML에서 class='text'처럼 지정된 class 이름으로 요소 찾기 -> 가장 자주 사용하는 방식
# 주의: class에 공백이 포함된 경우 -> ex) class='a b' 는 CLASS_NAME으로 바로 찾기 어렵다. -> 공백을 기준으로 구분함
# =============================================

#first_quote_text=driver.find_element(By.CLASS_NAME,'quote')
#print("\n CLASS_NAME으로 찾은 첫번째 명언 : ",first_quote_text.text)



# =============================================
# 3) id로 요소 찾기(ID)
# ID는 페이지에서 '유일하게' 존재 -> 정확하게 찾고 싶을 때 사용
# =============================================


#id_text=driver.find_element(By.ID,'ad-timeboard-response')
#print("\n ID로 찾은 요소의 태그 이름 : ",id_text.tag_name)


# =============================================
# 4) find_element vs find_elements
# find_element: 요소 1개만 찾기 (없으면 에러)
# find_elements: 요소 여러 개 찾고 리스트로 반환 (없으면 빈 리스트)
# =============================================


one_quote_block=driver.find_element(By.CLASS_NAME,'quote')
many_quote_block=driver.find_elements(By.CLASS_NAME,'quote')
print('\n find_element로 찾은 quote 블록 1개',type(one_quote_block))
print('\n find_element로 찾은 quote 블록 개수',len(many_quote_block))

# =============================================
# 5) xPath : HTML의 구조(경로)를 이용해서 요소를 찾는 방법
# ID 나 Class로 찾기 어려울 때 사용 , HTML 트리 구조를 따라가면서 요소를 지정할 수 있다.

# 장점: 조건이나 경로를 정확하게 요소 선택 가능
# 단점: 페이지 구조가 바뀌면 xPath도 같이 깨질 수 있다.
# ========================================== ===


# 첫 번째 명언 텍스트를 xPath로 찾기

''' 
xPath_quote_text=driver.find_element(
    By.XPATH,
    "//div[@class='quote']/span[@class='text']"   # => div 태그에 있는 class = 'quote' , 그 밑에 자식 class='text'
)
'''

'''
xPath_quote_text=driver.find_element(
    By.XPATH,
    "/html/body/div/div[2]/div[1]/div[1]/span[2]/small"   # 개발자도구로 검색한 영역 마우스 우클릭으로 Full xPath 복사
)


print("첫번째 위인:",xPath_quote_text.text)
'''





# 첫 번째 quote 블록만 특정해서 찾는 xPath 예시
xPath_first_block= driver.find_element(By.XPATH,"(//div[@class='quote'])[1]")
xPath_first_author= driver.find_element(By.CLASS_NAME,'author')
print("\n\n첫번째 명언: ",xPath_first_block.text)
print("\n\n xPath로 찾은 첫 번째 저자 : ",xPath_first_author.text)

# 브라우저 종료
driver.quit()
```
