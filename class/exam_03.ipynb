```python
# 클래스 속성

class Person():
    folder=[]   # 인스턴스 변수  -> 원래 인스턴스마다 변수를 다르게 가지지만, 
                # 클래스 속성  -> 클래스 속성은 모든 인스턴스에서 접근이 가능하고 공공제로 사용된다.

    def add_files(self,file):
        Person.folder.append(file) # 클래스 이름으로 클래스 속성에 접근




houng=Person()
houng.add_files("txt")
paul=Person()
paul.add_files("imag3131e")


print(houng.folder)
print(paul.folder)
```

```python
# 인스턴스 속성

class Person():
    def __init__(self):
        self.folder=[]
    
    def add_files(self,file):
        self.folder.append(file)


houng=Person()
houng.add_files("txt")
paul=Person()
paul.add_files("imag3131e")


print(houng.folder)
print(paul.folder)
```
