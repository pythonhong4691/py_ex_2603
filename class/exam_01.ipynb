# 클래스 -> 객체를 표현하기 위한 문법의 일종              
# 속성 -attribute
# 능력 (기능) -method 

# 객체가 객체를 지칭할 때 -> 객체
# 객체가 클래스와 연관지어서 지칭할 때 -> 인스턴스

# 

```python
class Person:
    def Hello(self): 
        print("Hello")
    def hi(self):
        self.Hello()

houng=Person() # houng은 Person 인스턴스 변수
houng.Hello() # 인스턴스 메서드
houng.hi() 
```
