# 속성, 기능

# __init__는 파이썬 클래스의 생성자 메서드
# 언더스코어 두 개(__)는 파이썬이 특별한 의미를 부여하는 예약된 패턴
# 따라서 init만 쓰면 단순한 메서드, __init__이어야 생성자로 동작합니다

```python
# 인스턴스 속성 사용
class Person():
    def __init__(self,name,age,hobby):
        self.hello="안녕하세요"     # 인스턴스 변수
        self.name=name
        self.age=age
        self.hobby=hobby
        
    def introduce(self):
        print(f"{self.hello},저는 {self.name}이고, {self.age}살입니다. 취미는 {self.hobby}입니다.")




houng=Person("호웅",25,"축구")
houng.introduce()


# print(f"{houng.age}")
# print(f"{houng.name}")
# print(f"{houng.hobby}")

        
```

```python
# 클래스 내부에서만 사용 가능한 
class Person():
    def __init__(self,name,age,hobby,weight):
        self.hello="안녕하세요"     # 인스턴스 변수
        self.name=name
        self.age=age
        self.hobby=hobby
        self.__weight=weight    # 비공개 속성
    
    def diet(self,loss_weight):
        self.__weight-=loss_weight
        print("현재 몸무계: ",self.__weight)
        
    def introduce(self):
        print(f"{self.hello},저는 {self.name}이고, {self.age}살입니다. 취미는 {self.hobby}입니다.")




houng=Person("호웅",25,"축구",90)
houng.introduce()
houng.diet(10)



# print(f"{houng.age}")
# print(f"{houng.name}")
# print(f"{houng.hobby}")

        
```
