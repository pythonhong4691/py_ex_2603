# 웹페이지에 데이터 추출

```python
# requests 모듈 -> 웹 페이지를 요청하고 응답 데이터를 받을 수 있음
import requests as req
from bs4 import BeautifulSoup  # html 해석기 (파이썬이 html 코드를 알 수 있게 해주는 모듈)

response=req.get('https://www.naver.com/')
print(response)  # Response [200] 로 요청 성공


html_doc = """<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>

<p class="story">Once upon a time there were three little sisters; and their names were
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>,
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a> and
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>;
and they lived at the bottom of a well.</p>

<p class="story">...</p>
"""


# print(html_doc)
# soup을 이용해서 html_doc 문자열 파싱, 그 내용을 soup 변수에 대입
soup=BeautifulSoup(html_doc,'html.parser') 

a_list=soup.select('a') # a태그들만 모아서 파싱



#print("===a 태그 리스트들===")
#print(a_list)


print("===a 태그 리스트 중 첫번째===")
print(a_list[0])

print("===a 태그 리스트 중 첫번째 내용===")
print(a_list[0].text) 

print("===a 태그 리스트 중 첫번째 링크===")
print(a_list[0]['href'])    # 속성을 선택할 땐 대괄호 -> ['~~']

print("\n===a 태그 리스트에 있는 텍스트 모두 출력===")

for a in a_list:
    print(a)

print("\n===a 태그 리스트에 있는 링크 모두 출력===")
for a in a_list:
    print(a['href'])



```

```python
import requests as req
from bs4 import BeautifulSoup  # html 해석기 (파이썬이 html 코드를 알 수 있게 해주는 모듈)


response=req.get("https://news.naver.com/")
#print(response)
#print(response.text)

soup=BeautifulSoup(response.text,'html.parser')


content=soup.select('.cnf_news_title')

print(content[0].text)


print("== class로 선택, 반복문으로 텍스트 출력 ==\n\n")
for c in content:
    print(c.text)


print("== \n\n 자식 선택자로 선택, 반복문으로 이미지 주소 출력 ==\n\n")
imgs=soup.select('.cnf_news_thumb img')
for i in imgs:
    print(imgs['data-src']) # print(img['src])


```
