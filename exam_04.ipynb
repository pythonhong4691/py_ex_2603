# 문자열[start:end(end-1)]

```python
s="안녕하세요"
print("s의 길이 출력")
print(len(s))

print("s를 이루는 문장 중에서 앞에 거, 2개 출력")
print(s[0:2])
print(s[1:2])
print(s[:4])
print(s[3:len(s)])
print(s[3:])
```
