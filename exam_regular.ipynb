# 정규식 : 문자열에서 특정 문자를 찾는 방식

```python
text :str="안녕하세요, 저는 25살 입니다."
print("25" in text)
```

```python
# 파이썬에서 정규표현식을 사용하려면 re모듈을 임포트
import re

text :str="안녕하세요, 저는 15살 입니다, 저는 안경이 15개 있습니다"

# r은 row string (특수 문자)
# \d는 숫자
# +는 1개 이상

pattern=r"\d+"     # 찾고 싶은 규칙
# re.findall(찾을 규칙, 찾을 문자열) -> 문자열에서 패턴과 맞는 내용들을 전부 찾아서 리스트로 반환
result=re.findall(pattern,text)

print(result)

```

```python
# 패턴 찾기
sample='abc123xyz456'

re.findall(r'\d+',sample)
re.findall(r'[a-z]+',sample)
```

```python
# 특정 형식 찾기

data='test123@gmail.com'
# \w : 문자 또는 숫자
# + : 1개 이상
# @ : 말 그래도 @
# w. : 점(특수 문자이기에 |를 붙인다.)
# 문자/숫자 + @ + 문자/숫자 + . +https://www.youtube.com/watch?v=Eh0Yo66ANUk
email_pattern=r'\w+@\w+\.\w+'
emails=re.findall(email_pattern,data)
print(emails)
```

```python
# 문자열 치환

test ="가격은 1200원 입니다."

# \d : 숫자
# sub('찾을 패턴',바꿀 문자, 대상 문자열)
# 숫자 문자열을 ''로 바꾼다.
new_text=re.sub(r'\d','',test)
print(new_text)
```

```python
# 정규식 vs 일반 문자열 처리

test = "몸무계 : 70kg, 키 :170cm"

test.split(',')
nums=re.findall(r'\d+',test)
print(nums)
```
